```python
import json
import pandas as pd
```

```python
# Load JSON data into a Python dictionary
with open('soa.json', 'r') as file:
    data = json.load(file)

# Convert the dictionary into a pandas DataFrame
# df = pd.json_normalize(data['Schedule of Activities'])

table_data = []
for entry in data["Schedule_of_Activities"]:
    for time_point in entry["Time_Post_Initial_Treatment"]:
        row = {"Day": entry["Day"], "Time_Post_Initial_Treatment": time_point}
        row.update(entry["Assessments"])
        table_data.append(row)

# Create a DataFrame from the list of dictionaries
df = pd.DataFrame(table_data)

# Display the DataFrame
#print(df)
df.head(n=100)
```
